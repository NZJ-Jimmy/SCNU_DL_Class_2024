```python
import pandas as pd

SGD = pd.read_csv("SGD_lr0.001_gamma0.8_acc.csv", index_col=0)
ADAM = pd.read_csv("ADAM_lr0.001_gamma0.8_acc.csv", index_col=0)
import matplotlib.pyplot as plt

# plt.figure(figsize=(10, 6))
plt.plot(SGD['epoch'], SGD['accuracy'], label='SGD', marker='o')
plt.plot(ADAM['epoch'], ADAM['accuracy'], label='ADAM', marker='x')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('SGD vs ADAM Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
